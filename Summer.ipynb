### 1. In how many cities Summer Olympics is held so far?

```python
# variable to hold the count
count = 0
# list to hold city values
cities = []
for i in range(0, len(df['City'])):
    
    if df['City'][i] not in cities: 
        
        cities.append(df['City'][i])
          
        count += 1
        
print("No.of cities :", count)  
print("Cities :", cities)
```

### 2. Which sport is having most number of Gold Medals so far? (Top 5)

```python
#to count medals
gold_cnt = 0
#to hold sports names
sports = []
#sports count
sp_cnt = 0
for i in range (0, len(df['Sport'])):
    if df['Sport'][i] not in sports:
        sports.append(df['Sport'][i])
        sp_cnt += 1
print("No.of sports :", sp_cnt)  
print("Sports :", sports)
```

```python
import collections
df.sort_values('Medal')
Gold_sport = []
occurrences = []
for i in range (0, len(df['Medal'])):
    if df['Medal'][i] == 'Gold' :
        Gold_sport.append(df['Sport'][i])
        gold_cnt += 1
occurrences = collections.Counter(Gold_sport)
print(occurrences)
```

```python
import matplotlib.pyplot as plt
x = [1421,1215,890,820,552]
y = ['Aquatics','Athletics','Rowing','Gymnastics','Fencing']
#plt.plot(x, y)
plt.plot(x, y, color='green', linestyle='dashed', linewidth = 3,
         marker='o', markerfacecolor='blue', markersize=12)
# naming the x axis
plt.xlabel('Gold count')
# naming the y axis
plt.ylabel('Sport')
plt.title('Top 5 sports with maximum Gold medals')
plt.show()
print('''
Aquatics: 1421
Athletics: 1215
Rowing: 890
Gymnastics: 820
Fencing: 552
''')
```

### 8. Which sport is having most female Gold Medalists? (Top 5)

```python
women_gold = 0
women_sports=[]

for i in range (0, len(df['Sport'])):
    if df['Gender'][i] == 'Women' and df['Medal'][i] == 'Gold':
        if df['Sport'][i] not in women_sports:
            women_sports.append(df['Sport'][i])
            women_gold += 1
occurrences = collections.Counter(women_sports)
print(occurrences)
```
